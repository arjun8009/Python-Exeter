```python
def find_max_state(dataset):
    state_dict = dict()
    data = dataset.groupby('hqstate')
    states = dataset['hqstate'].unique()
    for i in states:
        state_dict[i]=0
    for i in states : 
        for j in data.get_group(i)['totshequity']:
            state_dict[i] = state_dict.get(i)+j
    return state_dict
        
```

```python
value = find_max_state(dataset)
max(value, key=value.get)
```
