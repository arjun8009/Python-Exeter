```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import community as louvain_community
import matplotlib.cm as cm
import matplotlib.colors as color
```

```python
def draw_cluster(G,choice):
    layout = nx.spring_layout(G)
    clusters = nx.algorithms.community.louvain_communities(G)
    #print(clusters)
    node_color = [0 for  i in range(len(G.nodes()))]
    for i in range(len(clusters)):
        for j in list(clusters[i]):
            node_color[j] = i
    #print(node_color)
    #cmap = cm.get_cmap('plasma', len(clusters))
    nx.draw_networkx_nodes(G, layout, G.nodes(), node_size=140,
                       cmap=plt.cm.plasma, node_color=list(node_color))
    nx.draw_networkx_edges(G, layout, alpha=0.5)
    nx.draw_networkx_labels(G,layout,font_color='white')
    #plt.set_cmap(cmap)
    sm = plt.cm.ScalarMappable(cmap=plt.cm.plasma)
    plt.colorbar(sm)
    plt.show()
```

```python
G = nx.karate_club_graph()
draw_cluster(G,None)
```
