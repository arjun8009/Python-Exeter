```python
def objective_1(x,y):
    return x**3 + y**2

def objective_2(x,y):
    return y**2 - 4*x
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(0,10,0.1)
y = np.arange(0,10,0.1)
objective_vectors_1 = []
objective_vectors_2 = []
for i in range(0,len(x)):
    val1 = objective_1(x[i],y[i])
    val2 = objective_2(x[i],y[i])
    objective_vectors_1.append(val1)
    objective_vectors_2.append(val2)
plt.scatter(objective_vectors_1,objective_vectors_2)
plt.xlabel('Objective 1')
plt.ylabel('objective 2')
plt.show()
```

```python
plt.scatter(objective_vectors_1,objective_vectors_2)
plt.xlabel('Objective 1')
plt.ylabel('objective 2')
plt.xlim(50,150)
plt.ylim(0,10)
plt.show()
```

```python
weight_vectors_1 = np.arange(0,1,0.01)
weight_vectors_2 = 1-weight_vectors_1
origin = [0 for i in range(0,20)]
for i in range(0,10):
    plt.plot([0,weight_vectors_1[i]],[0,weight_vectors_2[i]])
plt.xlabel('Objective 1')
plt.ylabel('objective 2')
plt.show()
```

```python
weighted_vector_mul_1 = []
weighted_vector_mul_2 = []
for i in range(0,len(objective_vectors_1)):
    weighted_vector_mul_1.append(objective_vectors_1[i]*weight_vectors_1[0])
    weighted_vector_mul_2.append(objective_vectors_2[i]*weight_vectors_2[0])
weighted_sum=np.add(weighted_vector_mul_1,weighted_vector_mul_2)
```
